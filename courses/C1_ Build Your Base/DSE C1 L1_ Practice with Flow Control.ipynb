# L1 Q2:

Complete the Fibonacci sequence up to the 10th place using any or all of the following:
* `for` 
* `if` `elif` `else`
* `append`
* `range`

The Fibonacci sequence with indices:

1. 0
2. 1
3. 1
4. 2
5. 3
6. 5
7. 8
8. etc..

Example input:

`index = 10`

Example output:
```
Index: 10
Value: 34
```

````python
index = 10
for i in range(index):
  if i == 0:
    fib = [0]
  elif i == 1 or i == 2:
    fib.append(1)
  else:
    fib.append(fib[-1] + fib[-2])
print("Index: {}\nValue: {}".format(index,fib[-1]))
````
